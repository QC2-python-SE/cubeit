```python
import sys
import os
sys.path.append(os.path.abspath("../src"))
import cubeit
from cubeit.register import(
_QuantumRegister as QuantumRegister, 
QuantumState)
from cubeit.gates import(   
had, s, t, x, y, z,
cnot, cnot_10, swap, cz)
import cubeit.visualisation as vs
import numpy as np
import matplotlib.pyplot as plt
from cubeit.DMs import DensityMatrix2Qubit as DM
```

```python
from cubeit.gates_DMs import had, s, t, x, y, z, cz, cnot

state = (1/np.sqrt(2))*np.array([1.,0.,1.,0.], dtype=complex)
rho_00 = np.outer(state,state)
rho_00
state = DM(rho_00)
state.apply_sequence([had(), s(), t(), x(), cnot()], [0, 1, 1, 0, [1,0]])
state.apply_single_qubit_gate(had(), 1)
state.apply_single_qubit_gate(t(), 1)
state.apply_sequence([cz()], [[0,1]])

fig = vs.plot_circuit(state)
```

```python
system = QuantumRegister(2)
system.x(1)  # |01
system.had(0)  # (|0 + |1)/2  |1
system.swap(0, 1)  # |1  (|0 + |1)/2
system.cz(0, 1)  # Apply CZ gate
system.t(1)
system.s(0)
system.y(1)
system.z(0)
system.cnot(0, 1)  # Control on qubit 0, target on qubit 1
# Plot the recorded history as a circuit diagram
vs.plot_circuit(system)
```
